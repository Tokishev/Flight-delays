### Linear Regression fit with just dep_delay and get score

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
data_test = flights[["dep_delay","arr_delay"]]
```

```python
data_test.corr()
```

```python
data_test.shape
```

```python
data_test.isna().sum()
```

```python
data_test = data_test.dropna()
```

```python
data_test.shape
```

```python
X = data_test.dep_delay.to_numpy().reshape(-1,1)
y = data_test.arr_delay
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.7)
```

```python
clf = LinearRegression()
model = GridSearchCV(clf,param_grid={})
model.fit(X_train,y_train)
r2_score(y_test,model.predict(X_test))
```

```python
flights.loc[:,["fl_date","crs_arr_time","crs_dep_time"]] = flights[["fl_date","crs_arr_time","crs_dep_time"]].astype('string')

arr_time = pd.to_datetime(flights["fl_date"]+" "+ flights["crs_arr_time"].str.zfill(4).replace({'2400':'0000'}))
dep_time = pd.to_datetime(flights["fl_date"]+" "+ flights["crs_dep_time"].str.zfill(4).replace({'2400':'0000'}))

flights["arr_date_time"] = arr_time
flights["dep_date_time"] = dep_time

flights['arr_month'] = flights['arr_date_time'].dt.month
```

```python
# flights_modified['arr_year'] = flights_modified['arr_date_time'].dt.year
flights['arr_month'] = flights['arr_date_time'].dt.month
flights['arr_day_of_week'] = flights['arr_date_time'].dt.day_of_week
flights['arr_hour'] = flights['arr_date_time'].dt.hour


# flights_modified['dep_year'] = flights_modified['dep_date_time'].dt.year
flights['dep_month'] = flights['dep_date_time'].dt.month
flights['dep_day_of_week'] = flights['dep_date_time'].dt.day_of_week
flights['dep_hour'] = flights['dep_date_time'].dt.hour
```

```python
timing_plots = flights[["arr_month","arr_day_of_week","arr_hour","dep_month","dep_day_of_week","dep_hour","arr_delay"]]
```

```python
timing_plots
```

```python
sns.set_theme()
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 2.5})
a = timing_plots.groupby(by=["arr_month"])["arr_delay"].mean().reset_index().rename(columns={'arr_month':'month'})
a["type"] = "arrival"
b = timing_plots.groupby(by=["dep_month"])["arr_delay"].mean().reset_index().rename(columns={'dep_month':'month'})
b["type"] = "dep"
comb = pd.concat([a,b])
sns.lineplot(data=comb,x="month",y="arr_delay",hue="type",markers=True,marker=".",markersize=20)
plt.savefig('monthly delay pattern.jpeg')
plt.show()
```

```python
sns.set_theme()
a = timing_plots.groupby(by=["arr_hour"])["arr_delay"].mean().reset_index().rename(columns={'arr_hour':'hour'})
a["type"] = "arrival"
b = timing_plots.groupby(by=["dep_hour"])["arr_delay"].mean().reset_index().rename(columns={'dep_hour':'hour'})
b["type"] = "dep"
comb = pd.concat([a,b])
sns.lineplot(data=comb,x="hour",y="arr_delay",hue="type",markers=True,marker=".",markersize=20)
plt.savefig('hourly delay pattern.jpeg')
plt.show()
```

```python
sns.set_theme()
a = timing_plots.groupby(by=["arr_day_of_week"])["arr_delay"].mean().reset_index().rename(columns={'arr_day_of_week':'day_of_week'})
a["type"] = "arrival"
b = timing_plots.groupby(by=["dep_day_of_week"])["arr_delay"].mean().reset_index().rename(columns={'dep_day_of_week':'day_of_week'})
b["type"] = "dep"
comb = pd.concat([a,b]).replace({
    'day_of_week': {
        0 : 'Monday',
        1 : 'Tuesday',
        2 : 'Wednesday',
        3 : 'Thursday',
        4 : 'Friday',
        5 : 'Saturday',
        6 : 'Sunday'
    }
})
sns.lineplot(data=comb,x="day_of_week",y="arr_delay",hue="type",markers=True,marker=".",markersize=20)
plt.savefig('weekly delay pattern.jpeg')
plt.show()
```
