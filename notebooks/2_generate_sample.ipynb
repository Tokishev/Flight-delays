# Sample size reduction

```python
flights = flights_master.copy()
```

```python
flights.loc[:,["fl_date","crs_arr_time","crs_dep_time"]] = flights[["fl_date","crs_arr_time","crs_dep_time"]].astype('string')

arr_time = pd.to_datetime(flights["fl_date"]+" "+ flights["crs_arr_time"].str.zfill(4).replace({'2400':'0000'}))
dep_time = pd.to_datetime(flights["fl_date"]+" "+ flights["crs_dep_time"].str.zfill(4).replace({'2400':'0000'}))

flights["arr_date_time"] = arr_time
flights["dep_date_time"] = dep_time

flights['arr_month'] = flights['arr_date_time'].dt.month
```

```python
def reduce_size(data):
    return data.sample(frac=0.1,random_state=1)
flights_sampled = flights.groupby(by=["op_unique_carrier","arr_month","dest_airport_id"]).apply(reduce_size).reset_index(drop=True)
```

```python
flights_sampled.shape
```

```python
flights_sampled = flights_sampled.drop(columns=["Unnamed: 0"])
```
